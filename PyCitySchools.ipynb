```python
# Dependencies and Setup
import pandas as pd
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete

# To find the students who passed math with a score of 70+
pass_math= school_data_complete.loc[(school_data_complete["math_score"] >= 70)]

# To find the number of students who passed math with a score of 70+
stu_pass_math= pass_math["math_score"].count()
stu_pass_math

# To find the number of students who took the math exam
total_stu_math= school_data_complete["math_score"].count()
total_stu_math

# To find the percent of students who passed math with a score of 70+
perc_pass_math= (stu_pass_math/total_stu_math)*100
perc_pass_math

# To find the students who passed reading with a score of 70+
pass_read= school_data_complete.loc[(school_data_complete["reading_score"] >= 70)]

# To find the number of students who passed reading with a score of 70+
stu_pass_read= pass_read["reading_score"].count()
stu_pass_read

# To find the number of students who took the reading exam
total_stu_read= school_data_complete["reading_score"].count()
total_stu_read

# To find the percent of students who passed with reading score of 70+
perc_pass_read= (stu_pass_read/total_stu_read)*100
perc_pass_read

# To find the number of students who passed math and reading
total_pass= (stu_pass_math + stu_pass_read)
total_pass

# To find the number of students who took the math and reading exams
total_mr= (total_stu_math + total_stu_read)
total_mr

# To find the percent of overall pass in math and reading
perc_overall= (total_pass/total_mr)* 100
perc_overall

# To find the total number of schools 
schools= school_data_complete["school_name"].unique()
schools
total_schools= school_data_complete["school_name"].value_counts()
total_schools2= total_schools.count()
total_schools2

# To find the total students for the schools
total_students= school_data_complete["Student ID"].count()
total_students

# To find the total budget for the schools
school_budget= school_data_complete["budget"].unique()
total_budget= school_budget.sum()
total_budget

# To find the average math and reading scores
avg_math= school_data_complete["math_score"].mean()
avg_reading= school_data_complete["reading_score"].mean()
```

```python
# To create a district summary data frame
district_summary_df= {"Total Schools": [total_schools2],"Total Students": [total_students],\
                      "Total Budget": [total_budget], "Average Math Score": [avg_math], "Average Reading Score": [avg_reading],"% Passing Math": [perc_pass_math], "% Passing Reading": [perc_pass_read], "% Overall Passing": [perc_overall]}
district_summary_df
district_summary= pd.DataFrame(district_summary_df)

# To format the District Summary data frame
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,}".format)
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary.head()
```

```python
# School Summary 
# To find the type of each school
school_type= school_data_complete.groupby(["school_name"])["type"].unique().apply(lambda x:x[0])
school_type
# School Name
school_name= school_data_complete["school_name"].unique()
# Total Students per School
students= school_data_complete.loc[:, ["Student ID", "school_name"]]
tot_stu_school= school_data_complete.groupby(["school_name"])["Student ID"].count()
tot_stu_school

# To find total school budget
tot_budg_school= school_data_complete.groupby(["school_name"])["budget"].unique().apply(lambda x:x[0])
tot_budg_school

# To find per student budget per school
budget_per_stu= tot_budg_school/tot_stu_school
budget_per_stu

# To find Average Math Score per school
ss_avg_math= school_data_complete.groupby(["school_name"])["math_score"].mean()
ss_avg_math

# To find Average Reading Score per school 
ss_avg_read= school_data_complete.groupby(["school_name"])["reading_score"].mean()
ss_avg_read

# To find % Passing Math per school 
# To calculate the number of students passing math per school
ss_pass_math= school_data_complete[(school_data_complete["math_score"] >= 70)]
ss_pass_math= ss_pass_math.groupby(["school_name"]).count()["student_name"]
# To calculate the percentage of students passing math per school
ss_perc_pass_math = ss_pass_math / tot_stu_school * 100

# To find % Passing Reading per school
# To find how many students pass reading with score of 70+
ss_pass_read= school_data_complete[(school_data_complete["reading_score"] >= 70)]
# To find how many students pass reading per school
ss_pass_read= ss_pass_read.groupby(["school_name"]).count()["student_name"]
# To find the percentage of students who passed reading per school
ss_perc_pass_read = ss_pass_read / tot_stu_school * 100

#To find % Overall Passing per school 
# To find number of students passing overall per school
ss_overall_pass= school_data_complete[(school_data_complete["reading_score"] >= 70)
                                              & (school_data_complete["math_score"] >= 70)]
ss_overall_pass= ss_overall_pass.groupby(["school_name"]).count()["student_name"]
# To find the overall passing % per school
ss_perc_overall = ss_overall_pass / tot_stu_school * 100
```
